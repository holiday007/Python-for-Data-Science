**Uses item similarity**

```python
import numpy as np
import pandas as pd
```

```python
column_names = ["user_id","item_id", "rating", "timstamp"]
df = pd.read_csv("u.data", sep="\t", names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on = "item_id")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("darkgrid")
```

```python
df.groupby("title")["rating"].mean().sort_values(ascending = False).head()
```

```python
df.groupby("title")["rating"].count().sort_values(ascending = False).head()
```

```python
ratings = pd.DataFrame(df.groupby("title")["rating"].mean())
ratings.head()
```

```python
# how many people rated the movie matters
ratings["num of ratings"] = df.groupby("title")["rating"].count()
```

```python
ratings.head()
```

```python
ratings["num of ratings"].hist(bins = 70)
plt.xlim([0,400])
```

```python
plt.figure(figsize=[10,7])
sns.distplot(ratings["rating"], bins=70)
```

```python
sns.jointplot(x="rating", y="num of ratings", data = ratings, alpha = 0.2)
```

More ratings, more likely to have high ratings

```python
# user's rating on movie
moviemat = df.pivot_table(index = "user_id", columns="title", values="rating")
moviemat.head()
```

```python
ratings.sort_values("num of ratings", ascending=False).head(10)
```

```python
# Star Wars & Liar Liar
starwars_user_ratings = moviemat["Star Wars (1977)"]
liarliar_user_ratings = moviemat["Liar Liar (1997)"]
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings, axis = 0)
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings, axis = 0)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=["Correlation"])
corr_starwars.head()
```

```python
# but weirdly
corr_starwars.sort_values("Correlation", ascending=False).head(10)
```

Lot's of unheard of

```python
# filter out movies with number of reviews lower than a certain threshold
corr_starwars = corr_starwars.join(ratings["num of ratings"]) # join is merge on index
corr_starwars.head()
```

```python
corr_starwars[corr_starwars["num of ratings"] > 100].sort_values("Correlation", ascending = False).head(10)
```

```python
# Liar Liar
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns = ["Correlation"])
```

```python
corr_liarliar = corr_liarliar.join(ratings["num of ratings"])
```

```python
corr_liarliar[corr_liarliar["num of ratings"]>100].sort_values("Correlation",ascending = False).head(10)
```

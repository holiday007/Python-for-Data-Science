### Multiplot on same canvas

```python
plt.subplot(1,2,1) # arguments are (num of rows, num of columns, which plot (left-right, top-bottom))
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

### Object Oriented Methods

```python
# create a figure object, figure object is some kind of an imaginary blank canvas
fig = plt.figure()
print(type(fig))

# adding axes to figure object
axes = fig.add_axes([.1,.1,.8,.8]) # past in a list [left, bottom, width, height], ranging from 0 to 1

# use the axes as base for plotting
axes.plot(x,y)

# details
plt.xlabel("X Label") # or axes.set_xlabel()
axes.set_ylabel("Y Label")
axes.set_title("Title") # or plt.title()
plt.show()
```

### Multi Plots on Same Canvas | Plots in Plots

But much more flexible compared to functional method

```python
# create a figure object 
fig = plt.figure()

# create a set of axes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4, 0.3])

axes.plot(x,y)
axes.set_title("Larger Plot")
axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

### Figure Size, Aspect Ratio & DPI

```python
# figsize - tuple or list
# dpi - dots per inch, pixels per inch
fig = plt.figure(figsize = (4,2), dpi = 100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

**With Subplots**

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(4,2), dpi = 100)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

### Legends

* legend outside plot
    - .legend(bbox_to_anchor = (1, 0.7))

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Step 1, set labels
ax.plot(x,x**2, label="X Squared")
ax.plot(x,x**3, label="X Cubed")

# Step 2, legend
# loc specifys location of legends - 0(best) -> automatically create an good legend
ax.legend(loc = 0)
```

### Customizations 

**Color**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# color - in the "color" parameter
# both common color names and RGB Hex value works

# line width - "linewidth" parameter, or"lw"

# transparency - "alpha"

# linestyle - "linestyle", or "ls", common values: "-", "--", "-.", ":", "steps"

# marker - "marker", common values: ".", "o", "*", "v", "<", ">", "^", "+"
# markersize - "markersize"
# markerfacecolor - color inside a marker
# markeredgewidth
# markeredgecolor

ax.plot(x,y, color = "black", lw = 10, alpha = 0.1, ls = "-")
ax.plot(x,x**3, color = "#AA8A99", linestyle = "steps", marker = "o", 
        markersize = 20, markerfacecolor = "yellow", markeredgewidth=3,
       markeredgecolor = "red") # Google RGB Hex Code
```

**More customizations on line and markers on Matplotlib Concept lecture note**

- Some example
    * edgecolor

**X,Y Limits**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = "black", lw = 2, ls = "-")

# notice pass in a list
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_csv("df1", index_col=0)
df2 = pd.read_csv("df2")
df1.head()
```

```python
df2.head()
```

**Plotting off a column**

```python
# histogram
df1["A"].hist(bins=30, edgecolor = "black") # calling matplotlib under the hood
```

```python
# if you import seaborn, it will run seaborn, looks nicer
import seaborn as sns
sns.set_style("darkgrid")
df1["A"].hist(bins = 30, edgecolor = "black")
```

```python
# or use .plot
df1["A"].plot(kind = "hist", bins = 30, edgecolor = "black")
```

```python
# or
df1["A"].plot.hist()
```

```python
df2.head()
```

**Plot off entire data frames**

```python
# area plot
df2.plot.area(alpha = 0.2)
```

heights (from the top of one color to another on the same x index) indicates relative value

```python
df2.plot.line()
```

Not very useful unless the index means something (indicates changes in certain direction like time)  
- Can check if there are crazy values

```python
df1.plot.line(alpha = 0.5, lw = 0.5)
```

```python
# or to see a column
df1.plot.line(y="B", figsize = (15,3), lw = 0.5)
```

```python
# barplot 
df2.plot.bar()
```

```python
# stacked barplot
df2.plot.bar(stacked = True)
```

```python
# scatter plot
df1.plot.scatter(x = "A", y= "B", c = "C", cmap = "coolwarm", figsize = (8,5))  # note the parameter, "c"
```

```python
# color comparison
df1.plot.scatter(x="A", y="B", color = "C")
```

```python
# size
df1.plot.scatter(x="A", y="B", s = df1["C"]*50, edgecolor = "black")
```

```python
# box plot
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=["a", "b"])
df.head()
```

```python
# bivariate data

# hexbin plot - like scatterplot but show more information of density of an area
df.plot.hexbin(x="a", y="b", figsize = (9,6), gridsize = 25, cmap = "coolwarm", 
               edgecolor = "white", linewidth = 0.01)
```

```python
# density plot
df2["a"].plot.kde()

# same as .density()
```

```python
df2.plot.density()
```

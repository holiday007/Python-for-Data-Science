```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# load a seaborn data set
tips = sns.load_dataset("tips")
tips.head()
```

### Distribution 

**distplot**

```python
sns.distplot(tips["total_bill"], kde = False, bins = 30)
```

**join plot** - joints distribution of two variables

```python
sns.jointplot(x="total_bill", y="tip", data = tips)
```

```python
# hex instead of scatter
sns.jointplot(x="total_bill", y="tip", data = tips, kind="hex")
```

```python
# scatterplot +  a regression line
sns.jointplot(x="total_bill", y="tip", data = tips, kind="reg")
```

**Pair Plot**  
Pair-wise correlation among all numeric columns

```python
sns.pairplot(tips)
```

```python
# colorized by catgory using "hue"
# choose palette (themes of colors) "palette"
sns.pairplot(tips, hue = "sex", palette="coolwarm")
```

**rug plot** - dash marks

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"])
```

What is kde (kernel estimated density)?
 - https://en.wikipedia.org/wiki/Kernel_density_estimation

```python
# kde plot
sns.kdeplot(tips["total_bill"])
```

### Categorical plots
**Bar plot**

- Useful arguments:
    * stacked = True: to stack up values of multiple categories 

```python
sns.barplot(x="sex", y="total_bill", data = tips)
```

```python
import numpy as np
```

```python
# change the aggregate function, aka height of the bars
# "estimator"
sns.barplot(x="sex", y="total_bill", data = tips, estimator=np.std)
```

**count plot** - bar plot that counts the number of categories

```python
sns.countplot(x="sex", data=tips)
```

**box plot** by categories

```python
sns.boxplot(x="day", y="total_bill", data = tips)
```

```python
# to categorize by one more level using "hue"
sns.boxplot(x="day", y="total_bill", hue = "smoker", data = tips)
```

**violin plot**

```python
sns.violinplot(x = "day", y = "total_bill", data = tips)
```

```python
# same nature with hue
# but you can split it - "split"
sns.violinplot(x = "day", y = "total_bill", data = tips, hue = "sex", split = True)
```

**strip plot**

```python
# basically scatter plots with discrete values
sns.stripplot(x="day", y="total_bill", data = tips)
```

```python
# jitter
sns.stripplot(x="day", y="total_bill", data = tips, jitter=False)
```

```python
# hue and split
sns.stripplot(x="day", y="total_bill", data = tips, hue = "sex", split = True)
```

**swarm plot** - similar to violin plot

```python
sns.swarmplot(x="day", y="total_bill", data = tips)
```

Can be computationally heavy

```python
# combining violin plot and swarm plot
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
```

**factor plot**: general expression of all these plots

```python
sns.factorplot(x="day", y="total_bill", data=tips, kind = "box")
```

### Matrix Plot

```python
flights = sns.load_dataset("flights")
flights.head()
```

For heat map to work, data should be in matrix form  
- Index name and column name matches up

```python
# "annot" -> annotate the value of the cell
# "cmap" -> color map 
sns.heatmap(tips.corr(), annot = True, cmap = "coolwarm")
```

```python
flights_pivot = flights.pivot_table(index = "month", columns = "year", values = "passengers")
flights_pivot
```

```python
# "linecolor" - line color
# "linewidth" - to see some separation
sns.heatmap(flights_pivot,cmap="magma", linecolor="white", linewidths=0.2)
```

**Cluster Map** - Use hierarchical clustering

```python
sns.clustermap(flights_pivot)
```

```python
# normalize values from zero to one
sns.clustermap(flights_pivot, standard_scale=1, cmap = "coolwarm")
```

```python
# or z-score (assuming normal distribution)
sns.clustermap(flights_pivot, z_score=1, cmap = "coolwarm")
```

### Grids

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris["species"].unique()
```

```python
sns.pairplot(iris) # notice how the grid works
```

**Pair Grid**

```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
```

```python
# or 
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap="coolwarm")
```

**Facet Grid**

```python
g = sns.FacetGrid(data = tips, col = "time", row = "smoker")
g.map(sns.distplot, "total_bill")
```

```python
# some plots require too arguments
g = sns.FacetGrid(data = tips, col = "time", row = "smoker")
g.map(sns.scatterplot, "total_bill", "tip") # pass in both arguments
```

### Regression PLot

**lm plot**: scatter plot with a regression line

```python
sns.lmplot(x="total_bill", y="tip", data = tips, 
           hue = "sex", markers=["o","v"]) # note number of markers match number of categories
```

```python
sns.lmplot(x="total_bill", y="tip", data = tips, 
           hue = "sex", markers=["o","v"], scatter_kws={"s":1000})
```

```python
# calling facet grid arguments directly
sns.lmplot(x="total_bill", y="tip", data = tips, col = "sex", row = "time")
```

```python
sns.lmplot(x="total_bill", y="tip", data = tips, col = "day", hue = "sex")
```

### Customization - Style & Color

**aspect ratio**: height/width

```python
# aspect ratio
sns.lmplot(x="total_bill", y="tip", data = tips, col = "day", hue = "sex", aspect = 0.6)
```

**plot size**

```python
# size of plot
sns.lmplot(x="total_bill", y="tip", data = tips, col = "day", hue = "sex", aspect = 0.6, size = 2)
```

**sns.set_style**

```python
sns.set_style("darkgrid")
sns.countplot(x="sex", data = tips,)
```

**remove spines**

```python
sns.set_style("ticks")
sns.countplot(x="sex", data = tips)
```

```python
sns.set_style("ticks")
sns.countplot(x="sex", data = tips)
sns.despine() # "top", "right", "left", "bottom"
```

**size** by plt

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x="sex",data=tips)
```

**scale and context**

overrides default parameters and change font scales depending on what you are creating the figure for

```python
# context - "paper", "notebook", "talk", "poster"
# "font_scale"  = 3 - 3 times the size of default
sns.set_context("poster", font_scale=1.5)
sns.countplot(x="sex", data=tips)
```

**palette and colors**

```python
# palette
sns.set_context("notebook")
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", palette="seismic")
```
